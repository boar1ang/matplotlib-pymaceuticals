```python
#Matplotlib Homework
%matplotlib notebook

#Import Python packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st

# Study data files
mouse_metadata = "data/Mouse_metadata.csv"
study_results = "data/Study_results.csv"

#Read in data files
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)
```

```python
#Merge tables with inner join
merged_table = pd.merge(mouse_metadata, study_results, on="Mouse ID")
merged_table.head(22)
```

```python
merged_table.columns
```

```python
merged_table.set_index("Mouse ID")
```

```python
rows_per_drug = merged_table["Drug Regimen"].value_counts()
rows_per_drug_df = rows_per_drug.to_frame()
rows_per_drug_df = rows_per_drug_df.rename(columns={"Drug Regimen" : "Number of Data Points"})
rows_per_drug_df.index.name="Drug Name"
rows_per_drug_df
```

```python
sorted = merged_table.groupby(["Mouse ID", "Sex"])
sorted.count()

```
